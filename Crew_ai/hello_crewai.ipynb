````python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
from crewai import Agent, Task, Crew, Process # crewai í•µì‹¬ í´ë˜ìŠ¤ë“¤
from dotenv import load_dotenv
import os

# 1. í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ(.envì—ì„œ openai_api_keyë¶ˆëŸ¬ì˜¤ê¸°)
load_dotenv()

# 2. Agent ìƒì„± : role, goal, backstory ì„¤ì •
agent = Agent(
    role="AI ì–´ì‹œìŠ¤í„´íŠ¸", # ì—ì´ì „íŠ¸ì˜ ì—­í• 
    goal="ì‚¬ìš©ìì—ê²Œ ê°„ë‹¨í•œ í™˜ì˜ ì¸ì‚¬ë¥¼ ì œê³µ", # ì—ì´ì „íŠ¸ê°€ ë‹¬ì„±í•  ëª©í‘œ
    backstory="ë‹¹ì‹ ì€ ì¹œì ˆí•œ AI ë¹„ì„œë¡œ, ì–¸ì œë‚˜ ì •ì¤‘í•˜ê³  ë„ì›€ì´ ë˜ëŠ” ì¸ì‚¬ë¥¼ ê±´ë„µë‹ˆë‹¤." # ì—ì´ì „íŠ¸ì˜ ì„±ê²©/ë°°ê²½
)
````

````python
# 3. Task ìƒì„± : description, expected_output, agent ì§€ì •
task=Task(
    description="ì‚¬ìš©ìì—ê²Œ í™˜ì˜ ì¸ì‚¬ í•œë§ˆë””ë¥¼ ì‘ì„±í•˜ì„¸ìš”.", # ì—ì´ì „íŠ¸ê°€ ìˆ˜í–‰í•  ì‘ì—… ë‚´ìš©
    expected_output="í•œ ì¤„ì˜ ë”°ëœ»í•œ í™˜ì˜ ë©”ì‹œì§€", # ê¸°ëŒ€í•˜ëŠ” ì¶œë ¥ ê²°ê³¼ë¬¼
    agent=agent # ì´ íƒœìŠ¤í¬ë¥¼ ìˆ˜í–‰í•  ì—ì´ì „íŠ¸
)
````

````python
# 4. Crew ìƒì„± : ì—ì´ì „íŠ¸ì™€ íƒœìŠ¤í¬ë¥¼ í¬ë£¨ë¡œ êµ¬ì„± (ìˆœì°¨ í”„ë¡œì„¸ìŠ¤ ì„¤ì •)
crew = Crew(
    agents=[agent], # í¬ë£¨ì— ì†í•œ ì—ì´ì „íŠ¸ ëª©ë¡ (ì—¬ê¸°ì„œëŠ” 1ê°œ)
    tasks=[task], # í¬ë£¨ê°€ ìˆ˜í–‰í•  íƒœìŠ¤í¬ ëª©ë¡
    process=Process.sequential, # ìˆœì°¨ì  ì§„í–‰ ë°©ì‹ìœ¼ë¡œ íƒœìŠ¤í¬ ì‹¤í–‰
    verbose=True # ì‹¤í–‰ ì¤‘ ìƒì„¸í•œ ë¡œê·¸ ì¶œë ¥ ì—¬ë¶€
)
````

````python
# 5. Crew ì‹¤í–‰ : kickoff() ë©”ì„œë“œë¡œ íƒœìŠ¤í¬ ìˆ˜í–‰ ì‹œì‘
result = crew.kickoff()

# 6. ì‹¤í–‰ ê²°ê³¼ ì¶œë ¥
print(f"ìµœì¢… ê²°ê³¼ : {result}")
````

````python
# 2. ì—ì´ì „íŠ¸ ì •ì˜ (ì—°êµ¬ì›, ì‘ê°€)
researcher = Agent(
 role="AI Researcher",
 goal="ë©€í‹° ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œì˜ ì£¼ìš” ì¥ì ì„ 3ê°€ì§€ ì¡°ì‚¬í•˜ì—¬ ì„¤ëª…í•©ë‹ˆë‹¤.",
 backstory="ë‹¤ë…„ê°„ AI íŠ¸ë Œë“œë¥¼ ì—°êµ¬í•´ì˜¨ ì „ë¬¸ê°€ì…ë‹ˆë‹¤."
)
writer = Agent(
 role="Technical Writer",
 goal="ì—°êµ¬ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê°„ëµí•œ ê²°ë¡ ì„ ì‘ì„±í•©ë‹ˆë‹¤.",
 backstory="ë³µì¡í•œ ì •ë³´ë¥¼ ì‰½ê²Œ ìš”ì•½í•˜ëŠ” ì‘ë¬¸ ì „ë¬¸ê°€ì…ë‹ˆë‹¤."
)
````

````python
# 3. íƒœìŠ¤í¬ ì •ì˜ (ì¡°ì‚¬ íƒœìŠ¤í¬, ì‘ì„± íƒœìŠ¤í¬)
research_task = Task(
description="ë©€í‹° ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œì˜ ì£¼ìš” ì¥ì  3ê°€ì§€ë¥¼ ì¡°ì‚¬í•˜ê³  ê° ì¥ì ì„ ê°„ëµíˆ ì„¤ëª…í•˜ì„¸ìš”.",
expected_output="3ê°€ì§€ ì¥ì ì— ëŒ€í•œ ê°„ë‹¨í•œ ì„¤ëª… (bullet point ëª©ë¡)",
agent=researcher
)
write_task = Task(
 description="ìœ„ ì¡°ì‚¬ ê²°ê³¼ë¥¼ ì°¸ê³ í•˜ì—¬, ë©€í‹° ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œì˜ ì¥ì ì— ëŒ€í•œ ì§§ì€ ê²°ë¡ ì„ ì‘ì„±í•˜ì„¸ìš”.",
 expected_output="3~4ë¬¸ì¥ìœ¼ë¡œ êµ¬ì„±ëœ ê²°ë¡  ë‹¨ë½",
 agent=writer,
context=[research_task] # ì´ì „ íƒœìŠ¤í¬ì˜ ê²°ê³¼ë¥¼ í™œìš©
)
````

````python
# 4. í¬ë£¨ ìƒì„± (ìˆœì°¨ì  í”„ë¡œì„¸ìŠ¤ ì„¤ì •)
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task,write_task],
    process=Process.sequential,
    verbose=True
)
````

````python
# 5. í¬ë£¨ ì‹¤í–‰ ë° ê²°ê³¼ ì¶œë ¥
crew_output = crew.kickoff()

# 6. ê° íƒœìŠ¤í¬ì˜ ê²°ê³¼ë¥¼ ìˆœì°¨ì ìœ¼ë¡œ ì¶œë ¥
for idx, task_output in enumerate(crew_output.tasks_output, start=1):
    print(f"\n[Task {idx} Output]\n{task_output}\n")
````

````python
class MyCustomDuckDuckGoTool(BaseTool):
    name:str = "DuckDuckGo Search Tool"
    description:str = "ì›¹ì—ì„œ ìµœì‹  ì •ë³´ë¥¼ ê²€ìƒ‰í•  ìˆ˜ ìˆëŠ” ë„êµ¬ì…ë‹ˆë‹¤."
    
    def _run(self, query:str) -> str:
        duckduckgo_tool = DuckDuckGoSearchRun()
        response = duckduckgo_tool.invoke(query)
        return response
    
# DuckDuckGo ê²€ìƒ‰ ë„êµ¬ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
search_tool = MyCustomDuckDuckGoTool()
````

````python
# ë‘ ì—ì´ì „íŠ¸ ì •ì˜ (ì •ë³´ ì¡°ì‚¬ì & ì¼ì • ê¸°íšì)
research_agent = Agent(
    role="ì •ë³´ ì¡°ì‚¬ì",
    goal="{place} ì—¬í–‰ì— í•„ìš”í•œ ìµœì‹  ì •ë³´ë¥¼ ì¡°ì‚¬í•˜ì—¬ ì œê³µí•©ë‹ˆë‹¤.",
    backstory="ì˜¨ë¼ì¸ ì •ë³´ ê²€ìƒ‰ì— ëŠ¥í†µí•œ ì—¬í–‰ ì •ë³´ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.",
    tools=[search_tool], # ì›¹ ê²€ìƒ‰ ë„êµ¬ ì¥ì°©
    verbose=True
)

planner_agent = Agent(
    role="ì—¬í–‰ ì¼ì • ê¸°íšì",
    goal="ì œê³µëœ ì •ë³´ë¥¼ í™œìš©í•´ ì™„ì„±ë„ ë†’ì€ {place} ì—¬í–‰ ì¼ì •ì„ ì‘ì„±í•©ë‹ˆë‹¤.",
    backstory="êµ­ë‚´ ì—¬í–‰ ì¼ì •ì„ ì—¬ëŸ¬ ì°¨ë¡€ ê¸°íší•œ ê²½í—˜ì´ í’ë¶€í•œ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.",
    verbose=True
)
````

````python
# ë‘ ê°œì˜ Task ì •ì˜ ë° ì—°ê²°
research_task = Task(
    description=(
        "{place} ì—¬í–‰ì„ ìœ„í•´ ì•Œì•„ì•¼ í•  í•µì‹¬ ì •ë³´ë¥¼ ì¡°ì‚¬í•˜ì„¸ìš”.\n"
        "{place}ì˜ ì¸ê¸° ê´€ê´‘ì§€ ëª©ë¡, ì§€ì—­ë³„ ë§›ì§‘ ì¶”ì²œ, ì´ë™ ì‹œ ìœ ìš©í•œ êµí†µ ì •ë³´ ë“±ì„ ìµœì‹  ìë£Œë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì •ë¦¬í•´ ì£¼ì„¸ìš”."
    ),
    agent=research_agent,
    expected_output="í•œêµ­ì–´ë¡œ ì‘ì„±ëœ {place} ì—¬í–‰ì— ëŒ€í•œ ìš”ì•½ ì •ë³´ ëª©ë¡"
)

planning_task = Task(
    description=(
        "ìœ„ì˜ ì¡°ì‚¬ ê²°ê³¼ë¥¼ ì°¸ê³ í•˜ì—¬ {place}ì—ì„œ {days}ì¼ ë™ì•ˆ ë¨¸ë¬´ëŠ” ì—¬í–‰ ì¼ì •ì„ ì‘ì„±í•´ ì£¼ì„¸ìš”.\n"
        "ê° ë‚ ì§œë³„ë¡œ ì˜¤ì „/ì˜¤í›„/ì €ë… ê³„íšì„ ì„¸ìš°ê³ , ì¡°ì‚¬ëœ ê´€ê´‘ì§€ì™€ ë§›ì§‘ ì •ë³´ë¥¼ ì¼ì •ì— ë°˜ì˜í•˜ì„¸ìš”.\n"
        "ì¼ì •ì—ëŠ” ë°©ë¬¸ì§€ì— ëŒ€í•œ ê°„ë‹¨í•œ ì„¤ëª…ì´ë‚˜ ì—¬í–‰ íŒë„ í¬í•¨í•´ ì£¼ì„¸ìš”."
    ),
    agent=planner_agent,
    context=[research_task], # ì´ì „ ì¡°ì‚¬ ê²°ê³¼ë¥¼ ì»¨í…ìŠ¤íŠ¸ë¡œ ì „ë‹¬
    expected_output="í•œêµ­ì–´ë¡œ ì‘ì„±ëœ ì¡°ì‚¬ëœ ì •ë³´ë¥¼ ë°˜ì˜í•œ {days}ì¼ê°„ì˜ ì—¬í–‰ ì¼ì •"
)
````

````python
# ë‘ ì—ì´ì „íŠ¸ë¥¼ Crewë¡œ ë¬¶ì–´ ìˆœì°¨ ì‹¤í–‰
crew_multi = Crew(
    agents=[research_agent, planner_agent],
    tasks=[research_task, planning_task],
    process=Process.sequential,
    verbose=True
)

print("\n=== [í˜‘ì—… ì—ì´ì „íŠ¸] í‰ì–‘ 3ì¼ ì¼ì • ìƒì„± ì‹œì‘ ===")
result_multi = crew_multi.kickoff(inputs={"place":"í‰ì–‘", "days":3})
print("=== [í˜‘ì—… ì—ì´ì „íŠ¸] ìƒì„±ëœ í‰ì–‘ 3ì¼ ì¼ì • ===")
show_full_output(result_multi)
````
