```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 클러스터 데이터 생성
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=0)

# K-평균 군집화 모델 생성 및 학습
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)

# 군집 레이블 출력
print('Labels :', kmeans.labels_)

# 군집 중심점 출력
print(f"Cluster centers : {kmeans.cluster_centers_}")

# 시각화
plt.figure(figsize=(10, 7))
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='r', marker='x')
plt.title("K-Means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 고객 데이터 생성(연령, 소득)
X = np.array([[35, 50000], [45, 85000], [25,35000],
              [60,90000], [50,80000], [35,55000]])

# K-평균 군집화 모델 생성 및 학습(k=3)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# 군집 레이블 출력
print(f"Labels : {kmeans.labels_}")

# 군집 중심점 출력
print(f"Cluster centers : {kmeans.cluster_centers_}")

# 시각화
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='r', marker='*', label='Centroids')
plt.title("K-Means Customer Segmentation")
plt.xlabel("Age")
plt.ylabel("Annual Income ($)")
plt.legend()
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

# 모든 행 출력
pd.set_option('display.max_rows', None)

# 모든 열 출력
pd.set_option('display.max_columns', None)

# 열 너비 제한 없애기
pd.set_option('display.max_colwidth', None)

# 전체 출력 시 생략 부호(...) 제거
pd.set_option('display.expand_frame_repr', False)

# 데이터 생성
# transactions = [
#     ['bread', 'milk'],
#     ['bread','jam','milk'],
#     ['beer','cola'],
#     ['bread','milk','cola'],
#     ['milk', 'cola','eggs']
# ]
transactions = [
    ['milk', 'bread', 'butter'],
    ['beer', 'diapers', 'chips'],
    ['milk', 'bread'],
    ['bread', 'butter'],
    ['beer', 'chips'],
    ['diapers', 'milk', 'bread'],
    ['milk', 'bread', 'butter', 'jam'],
    ['jam', 'butter'],
    ['diapers', 'beer'],
    ['milk', 'diapers', 'bread', 'beer'],
    ['cola', 'chips'],
    ['cola', 'milk', 'diapers'],
    ['bread', 'butter', 'chips'],
    ['jam', 'milk'],
    ['milk', 'bread', 'butter', 'chips', 'diapers']
]


# # 트랜잭션 데이터를 one-hot encoding으로 dataframe 변환
# df = pd.DataFrame(transactions)
# df = df.apply(pd.Series).fillna(0).astype(bool).astype(int)

# 원-핫 인코딩
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# 연관 규칙 학습
item_sets = apriori(df, min_support=0.2, use_colnames=True)
rules = association_rules(item_sets, metric='confidence', min_threshold=0.6)

print(df)

# 결과 출력
print(rules)
```

IsolationForest

- 이상치(Outlier)를 탐지하는 앙상블 기반 비지도 학습 모델
- Scikit-learn의 IsolationForest는 랜덤 트리로 데이터를 "격리"시키는 방식으로 작동
- 일반 데이터는 많은 분할이 필요하지만, 이상치는 적은 분할로도 쉽게 분리됨
- 이 아이디어를 바탕으로 이상치일 가능성을 점수화함

하이퍼파라미터: contamination
- contamination=0.1은 "전체 데이터 중 약 10%를 이상치로 간주하겠다"는 의미
- 이 값은 이상치 비율에 대한 '사전 추정치'로, 임계값 결정에 사용됨

예시 사용 흐름:

from sklearn.ensemble import IsolationForest

# 모델 생성
model = IsolationForest(contamination=0.1)

# 학습 (X는 feature 데이터프레임 또는 넘파이 배열)
model.fit(X)

# 예측 (-1: 이상치, 1: 정상)
pred = model.predict(X)

# 이상치일 확률을 추정한 score도 가능 (낮을수록 이상치 가능성 큼)
scores = model.decision_function(X)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 예제 데이터 생성
np.random.seed(42)

# X는 평균이 0이고 표준편차가 0.3(=1*0.3)인 정규분포를 따르는 100개의 2차원 데이터 포인트입니다.
X = 0.3 * np.random.randn(100,2)
X_train = np.r_[X+2, X-2] # NumPy의 r_ 함수는 여러 개의 배열을 row방향으로 concatenate(연결)

# X_outliers는 각 차원이 -4와 4사이의 균일 분포를 따르는 20개의 2차원 이상치 데이터 포인트입니다.
X_outliers = np.random.uniform(low=-4, high=4, size=(20,2))

# 전체 데이터 (정상 데이터 + 이상치)
X_total = np.r_[X_train, X_outliers]

# Isolation Forest 모델 생성 및 학습
model = IsolationForest(contamination=0.1)
model.fit(X_train)

# 이상치 예측
y_pred_train =  model.predict(X_train)
y_pred_outliers = model.predict(X_outliers)

# 결과 시각화
plt.scatter(X_train[:,0], X_train[:,1], c='white', edgecolors='k', s=20, label="Training Data")
plt.scatter(X_outliers[:,0], X_outliers[:,1], c='red', edgecolors='k', s=20, label="Outliers")
plt.legend()
plt.title("Isolation Forest for Anomaly Detection")
plt.show()
```

```python
import numpy as np
import pandas as pd

# 랜덤 데이터 생성 (정상 거래)
np.random.seed(42)
transaction_amounts = np.random.normal(100, 20, 1000) # 평균 100, 표준편차 20인 거래 금액 1000개
transaction_times = np.random.normal(15, 5, 1000) # 평균 15, 표준편차 5인 거래 시간 1000개

# 몇몇 이상치 추가
anomalous_amounts = np.random.uniform(300, 500, 50) # 300~500 사이의 이상치 거래 금액 50개
anomalous_times = np.random.uniform(30, 50, 50) # 30~50 사이의 이상치 거래 시간 50개

# 정상 거래와 이상치 데이터 합치기
amounts = np.concatenate([transaction_amounts, anomalous_amounts])
times = np.concatenate([transaction_times, anomalous_times])

# 데이터프레임 생성
df = pd.DataFrame({'amount': amounts, 'time' : times})
print(df)
```

```python
from sklearn.ensemble import IsolationForest

# 모델 생성 및 학습
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(df)

# 이상치 예측
df['is_outlier'] = model.predict(df)

# 이상치는 -1, 정상치는 1로 표시됨
df['is_outlier'] = df['is_outlier'].map({1:'normal', -1:'outlier'})

# 결과 출력
print(df['is_outlier'].value_counts())
print(df.head())
```

```python
from sklearn.neighbors import LocalOutlierFactor

# LOF 모델 생성 및 학습
lof = LocalOutlierFactor(n_neighbors=30, contamination=0.05)
df['lof_outlier'] = lof.fit_predict(df[['amount','time']])

# 이상치는 -1, 정상치는 1로 표시됨
df['lof_outlier'] = df['lof_outlier'].map({1:'normal', -1:'outlier'})

# 결과 출력
print(df['lof_outlier'].value_counts())
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Isolation Forest 결과 시각화
plt.figure(figsize=(10, 5))
plt.scatter(df['amount'], df['time'], c=df['is_outlier'].map({'normal': 'b', 'outlier': 'r'}))
plt.xlabel('Transaction Amount')
plt.ylabel('Transaction Time')
plt.title('Isolation Forest Outlier Detection')
plt.show()
# LOF 결과 시각화
plt.figure(figsize=(10, 5))
plt.scatter(df['amount'], df['time'], c=df['lof_outlier'].map({'normal': 'b', 'outlier': 'r'}))
plt.xlabel('Transaction Amount')
plt.ylabel('Transaction Time')
plt.title('Local Outlier Factor (LOF) Outlier Detection')
plt.show()
```
